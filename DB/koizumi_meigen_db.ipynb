```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

```python
#スクレ―ピングするための構文
# ページのURL
url = 'https://hikotore.com/koizumishinzirou/#index_id35'

# リクエストを送信してHTMLを取得
response = requests.get(url)
html = response.content
```

```python
# BeautifulSoupオブジェクトを作成
soup = BeautifulSoup(html, 'html.parser')

# HTMLの全体を表示
print(soup.prettify())
```

```python
# .is-style-num_circleクラスが含まれるリストアイテムの情報を抽出
list_items = soup.select('.is-style-num_circle li')
for item in list_items:
    print(item.get_text())
```

```python
# .is-style-num_circleクラスが含まれるリストアイテムの情報を抽出
list_items = soup.select('.is-style-num_circle li')

# テキストファイルに名言を保存（文字コードをutf-8に設定）
with open('quotes.txt', 'w', encoding='utf-8') as f:
    for item in list_items:
        f.write(item.get_text() + '\n')
```

```python
# ファイルを開く
with open('quotes.txt', 'r', encoding='utf-8') as f:
    # 一行ずつ読み込み、改行文字を削除
    quotes = [line.strip() for line in f]
```

```python
# データベースに接続（データベースが存在しない場合は新規作成）
conn = sqlite3.connect('../services/quotes_20240417_135122_加工用 - コピー.db')
```

```python
# カーソルを取得
c = conn.cursor()
```

```python
# 取得した情報をデータベースに追加
for quote in quotes:
    c.execute("INSERT INTO quotes (title, quote, author, url) VALUES (?, ?, ?, ?)",
              ('小泉進次郎の名言集', quote, '小泉進次郎', 'https://hikotore.com/koizumishinzirou/'))
```

```python
# 変更をコミット（保存）
conn.commit()
```

```python
# 名言を追加した後の最大のIDを取得
c.execute("SELECT MAX(id) FROM quotes")
max_id_after = c.fetchone()[0]

# 最大のIDを出力
print("最大のID（追加後）: ", max_id_after)
```

```python
# カーソルを取得
c = conn.cursor()

# 最新の5行を取得
c.execute("SELECT * FROM quotes ORDER BY id DESC LIMIT 5")

# 結果を取得
rows = c.fetchall()

# 結果を表示
for row in rows:
    print(row)
```

```python
# データベース接続を閉じる
conn.close()
```
